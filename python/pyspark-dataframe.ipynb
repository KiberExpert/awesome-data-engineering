```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
spark=SparkSession.builder.appName(name="DataFrame operations").getOrCreate()
```

```python
data=[(1, "Alice"), (2, "John"), (3, "Brain")]

df=spark.createDataFrame(data, ["Age", "Name"])

df.show()
```

```python
selected_columns=df.select('age')

selected_columns.show()
```

```python
filtered_data=df.filter('age==1')

filtered_data.show()
```

```python
filtered_data=df.filter(df['age']==1)

filtered_data.show()
```

```python
group_data=df.groupBy('Name').agg({"age":"sum"})
group_data.show()
```

```python
df2=df.select('age').limit(1)

join_data=df.join(df2, "age", "inner")

join_data.show()
```

```python
df2 = df.select('age').limit(1)
join_data = df.join(df2, "age", "left")
# join_data.filter(join_data['age'].isNotNull()).select('age')
join_data.show()
```

```python
join_data.orderBy('age').show()
```

```python
from pyspark.sql.functions import col, desc
join_data.orderBy((col("age").desc())).show()
```

```python
join_data.select('age').distinct().show()
```

```python
join_data2=join_data.withColumn('age2', join_data['age']+1)
join_data2.show()
```

```python
join_data2.drop(join_data2['age2'])

join_data2.show()
```

```python
join_data3=join_data2.withColumnRenamed('age2', 'age3')

join_data3.show()
```
