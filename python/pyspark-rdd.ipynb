```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
spark = SparkSession.builder\
        .appName(name='RDD operators')\
        .getOrCreate()

spark
```

```python
numbers = [1,2,3,4,5]

rdd = spark.sparkContext.parallelize(numbers)
```

```python
rdd.collect()
```

```python
data = [("Alice", 25), ("Bob", 30), ("Charlie", 35), ("Alice", 40)]
rdd=spark.sparkContext.parallelize(data)

rdd.collect()
```

```python
rdd.count()
```

```python
rdd.first()
```

```python
rdd.take(3)
```

```python
rdd.foreach(lambda x: print(x))
```

```python
mapped_rdd = rdd.map(lambda x: (x[0].upper(), x[1]))
mapped_rdd.collect()
```

```python
filter_rdd=rdd.filter(lambda x: x[1]>3)
filter_rdd.collect()
```

```python
reduce_rdd=rdd.reduceByKey(lambda x, y: x+y)
reduce_rdd.collect()
```

```python
sorted_rdd=rdd.sortBy(lambda x: x[1], ascending=True)

sorted_rdd.collect()
```

```python
spark.stop()
```
