```python
import pandas as pd
```

```python
# bo'sh qiymatlar bilan ishlash 

data = {'name': ['Ali', 'Vali', None, 'Sami'],
        'score': [85, None, 75, 90]}

df = pd.DataFrame(data)

df['score'] = df['score'].fillna(0)
df['name'] = df['name'].fillna('NoName')
print(df)
```

```python
# ustun qiymatlarini o'zgartirish
df['score'] = df['score'] * 2

print(df)
```

```python
# yangi ustun qo'shish
df['result'] = df['score'] / 200 
print(df)
```

```python
# shartli ravishda ustun yaratish 

df['level'] =  df['score'].apply(lambda x: int( x* 10) if x > 0 else 0 )

print(df)
```

```python
# sana va vaqt bilan ishlash 

try:
    data = {'name': ['Ali', 'Vali'],
            'date': ['2023-01-10', '2023-05-25']}
    
    df = pd.DataFrame(data)
    
    df['date'] = pd.to_datetime(df['date'])
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    
    print(df)
    
except Exception as e:
    print(e)
```

```python
# frame ustunlarini qayta nomlash 
df.rename({'name': 'ism', 'date': 'sana', 'month': 'oy', 'year': 'yil'}, inplace=True)
```

```python
# guruhlash va agregatsiya 
data = {'department': ['IT', 'HR', 'IT', 'HR'], 'salary': [500, 400, 600, 450]}

df = pd.DataFrame(data)

print(df.groupby('department')['salary'].sum())
```

```python
# saralash 
df = df.sort_values(by='salary', ascending=True)
print(df)
```

```python
# replace
data = {'gender': ['M', 'F', 'M', 'F']}

df = pd.DataFrame(data)

df['gender'] = df['gender'].map({'M': "Male", "F": "Female"})
df
```

```python
#rename
df.rename({'gender': 'jinsi'}, inplace=False)
```

```python
# split
data = {'email': ['ali@gmail.com', 'vali@yandex.com']}
df=pd.DataFrame(data)

df['domen'] = df['email'].str.split('@').str[1]

print(df)
```
