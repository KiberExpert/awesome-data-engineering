```python
import pandas as pd
```

```python
# Faqat Name ustunini ajratib oling.
df1 = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Sami'],
    'Age': [24, 30, 22],
    'City': ['Tashkent', 'Bukhara', 'Andijan']
})

print(df1['Name'])
```

```python
# Mahsulotlarning o‘rtacha narxini toping.
df2 = pd.DataFrame({
    'Product': ['Apple', 'Banana', 'Orange'],
    'Price': [1000, 2000, 1500]
})
print(df2['Price'].mean())
```

```python
# Har bir talabaga Total ustuni qo‘shing, u Math + English bo‘lsin.

df3 = pd.DataFrame({
    'Student': ['Aziz', 'Shox', 'Dildora'],
    'Math': [78, 85, 90],
    'English': [88, 75, 95]
})
df3['Total'] = df3['Math']+df3['English']
print(df3)
```

```python
# Har bir shahar bo‘yicha o‘rtacha aholini hisoblang.

df4 = pd.DataFrame({
    'City': ['Tashkent', 'Tashkent', 'Bukhara'],
    'Population': [2500000, 2600000, 500000]
})

print(df4.groupby('City')['Population'].mean())
```

```python
# Score ustunidagi NaN qiymatlarni 0 bilan to‘ldiring.

df5 = pd.DataFrame({
    'Name': ['Malika', 'Otabek', 'Nodir'],
    'Score': [None, 80, 95]
})
df5['Score'] = df5['Score'].fillna(0)
df5
```

```python
# Har bir ishchining umumiy daromadini (Salary + Bonus) hisoblang.
try:
    df6 = pd.DataFrame({
        'ID': [1, 2, 3],
        'Salary': [3000, 4000, 3500],
        'Bonus': [500, 700, 600]
    }) 
    print(df6.groupby('ID')['Salary'].sum()+df6.groupby('ID')['Bonus'].sum())
except Exception as e:
    print(e)
```

```python
# Score ≥ 80 bo‘lganlarni ajratib oling.
df7 = pd.DataFrame({
    'Name': ['A', 'B', 'C'],
    'Score': [90, 70, 80]
})
print(df7[df7['Score']>=80])
```

```python
# Date ustunini datetime formatiga o‘tkazing.
df8 = pd.DataFrame({
    'Date': ['2023-01-01', '2023-02-01', '2023-03-01'],
    'Sales': [100, 200, 150]
})
df8['Date']=pd.to_datetime(df8['Date'])
df8
```

```python
# Omborda mavjud mahsulotlarni ajratib oling.
df9 = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'InStock': [True, False, True]
})
print(df9[df9['InStock']==True])
```

```python
# A ustunini indeks sifatida belgilang.
df10 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df10 = df10.set_index('A')
df10
```

```python
# Qiymatlar yig‘indisini hisoblang.
df11 = pd.DataFrame({
    'Region': ['North', 'South', 'East'],
    'Value': [100, 150, 120]
})
df11.sum()['Value']
```

```python
# Yoshlari 25 dan kichiklarni tanlab oling.
df12 = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Sami'],
    'Age': [24, 30, 22]
})

df12[df12['Age']<25]
```

```python
# Har bir ism bo‘yicha o‘rtacha bahoni hisoblang.
df13 = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'A'],
    'Score': [80, 90, 85, 70]
})

df13.groupby('Name')['Score'].mean()
```

```python
# Faqat faol foydalanuvchilarni ajrating.
df14 = pd.DataFrame({
    'User': ['a', 'b', 'c'],
    'Active': [False, True, True]
})

df14[df14['Active']==True]
```

```python
# Harorat 30 dan yuqori bo‘lgan shaharlarni ajrating.
df15 = pd.DataFrame({
    'City': ['Tashkent', 'Andijan', 'Bukhara'],
    'Temp': [35, 28, 32]
})

df15[df15['Temp']>30]
```

```python
# Age ustunini kamayish tartibida tartiblash.
df16 = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Sami'],
    'Age': [24, 30, 22]
})
result = df16.sort_values(by='Age', ascending=False)
print(result)
```

```python
# A va B ustunlarining farqini hisoblang.
df17 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [3, 2, 1]
})

print(df17['A']-df17['B'])
```

```python
# Narxi 150 dan katta bo‘lgan itemlarni chiqaring.
df18 = pd.DataFrame({
    'Item': ['X', 'Y', 'Z'],
    'Price': [100, 200, 150]
})

df18[df18['Price']>150]
```

```python
# Har bir harfni ASCII kodi bilan birlashtirib yangi ustun yarating.
df19 = pd.DataFrame({
    'Letter': ['a', 'b', 'c'],
    'Code': [97, 98, 99]
})

df19['ADCII'] = df19['Letter']+df19['Code'].astype(str)
print(df19)
```

```python
# Eng ko‘p sotilgan mahsulotni aniqlang.
df20 = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Sold': [100, 200, 150]
})

print(df20[df20['Sold']==df20['Sold'].max()])
```

```python
# students va scores DataFrame'larini id va student_id orqali birlashtiring.
students = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Ali', 'Vali', 'Sami']
})
scores = pd.DataFrame({
    'student_id': [1, 2, 3],
    'score': [85, 90, 78]
})

print(pd.merge(students, scores, left_on='id', right_on='student_id'))
```

```python
# Har bir talabaning o‘rtacha bahosini hisoblang (takroriy ismlar bo‘yicha).
df = pd.DataFrame({
    'name': ['Ali', 'Vali', 'Sami', 'Ali'],
    'score': [80, 85, 90, 95]
})
print(df.groupby('name')['score'].mean())
```

```python
# query() funksiyasi yordamida umumiy balli 160 dan yuqori bo‘lganlarni ajrating.
df = pd.DataFrame({
    'name': ['Ali', 'Vali', 'Sami'],
    'math': [80, 85, 90],
    'eng': [70, 75, 85]
})

df.query('(math+eng)>160')
```

```python
# Faqat 2024-01-03 sanadan keyingi ma'lumotlarni tanlang.

df = pd.DataFrame({
    'date': pd.date_range('2024-01-01', periods=5, freq='D'),
    'value': [10, 20, 30, 40, 50]
})

df.query("date > '2024-01-03'")
```

```python
# status == 'active' bo‘lganlar ustida filter() funksiyasidan foydalaning.
df = pd.DataFrame({
    'user': ['a', 'b', 'c'],
    'status': ['active', 'inactive', 'active']
})

df.query("status=='active'").filter(items=['user'])
```

```python
# Har bir foydalanuvchi bo‘yicha jami buyurtma summasini hisoblang.
orders = pd.DataFrame({
    'order_id': [1, 2, 3, 4],
    'user_id': [101, 102, 101, 103],
    'amount': [250, 300, 150, 400]
})

orders.groupby('user_id')['amount'].sum()
```

```python
# Har bir mahsulot uchun yakuniy narxni hisoblang (Price - Discount%).

df = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Price': [1000, 2000, 1500, 3000],
    'Discount': [10, 20, 0, 15]
})

df['Salary']=df['Price']*df['Discount']
df
```

```python
# Har bir Group bo‘yicha Score ustunini normallashtiring (min-max scaling).

df = pd.DataFrame({
    'Group': ['X', 'Y', 'X', 'Y'],
    'Score': [10, 20, 30, 40]
})

df.groupby('Group')['Score'].sum()
```

```python
# LoginTime ustunini datetime formatiga o‘tkazing va eng ertaroq kirgan foydalanuvchini aniqlang.
df = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Sami'],
    'LoginTime': ['08:00', '09:15', '08:45']
})
df['LoginTime'] = pd.to_datetime(df['LoginTime'], format='%H:%M')
df['Name'].min()
```

```python
# left va right DataFrame’larini outer join qilib birlashtiring.
left = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Vali', 'Sami']
})
right = pd.DataFrame({
    'ID': [2, 3, 4],
    'Grade': ['B', 'A', 'C']
})

pd.merge(left, right, left_on='ID', right_on='ID')
```

```python

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [5, 6, 7, 8]
})
df.max(axis=1)
```

```python
# Manfiy balansga ega mijozlarni alohida DataFrame’da ajrating.

df = pd.DataFrame({
    'Customer': ['A', 'B', 'C'],
    'Balance': [1000, -50, 500]
})

df[df['Balance']<0]
```

```python
# A == 'foo' bo‘yicha guruhlab B ustunining o‘rtacha qiymatini toping.

df = pd.DataFrame({
    'A': ['foo', 'bar', 'foo', 'bar'],
    'B': [1, 2, 3, 4]
})


df.groupby('A')['B'].mean()
```

```python
# Ustunlar bo‘yicha (axis=0) va qatnashgan holda (axis=1) yig‘indilarni hisoblang.
df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

df.sum(axis=0)
```

```python
# Har bir foydalanuvchining hozirgi yoshini hisoblang (hozirgi yilni hisobga olib).
import datetime

df = pd.DataFrame({
    'Name': ['Ali', 'Vali', 'Sami'],
    'BirthYear': [2000, 1995, 1998]
})

df['Age']=datetime.date.today().year-df['BirthYear']
df
```

```python
# Faqat mavjud (Qty > 0) itemlar sonini hisoblang.
df = pd.DataFrame({
    'Item': ['A', 'B', 'C', 'D'],
    'Qty': [10, 0, 5, 3]
})

df[df['Qty']>0].groupby('Item')['Qty'].count()
```

```python
import pandas as pd

df = pd.DataFrame({
    'Value': [10, None, 30, None, 50]
})

df['Value'].fillna(df['Value'].mean(), inplace=True)
print(df)
```
